## Identify the output variable.

```python
feature=train.drop("ACTION",axis=1)
target=train["ACTION"]
```

## Predict the accuracy using classification models.

```python
from sklearn.ensemble import RandomForestClassifier
RFC=RandomForestClassifier()

from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100,150,200,250],
    'max_depth': [None,1,2,3],
    'min_samples_leaf': [3,4,5,6]
}

gsF = GridSearchCV(RFC, param_grid = param_grid, cv=3, verbose=0)
gsF.fit(feature,target)

print(gsF.best_params_)
acc_RFG=gsF.best_score_
acc_RFG
```

```python
predictions=gsF.predict(test)
```

### First let's see how the feature are correlated with eath other and with the target variable

```python
plt.figure(figsize=(20,15))
sns.heatmap( train.corr()  , cmap="YlGnBu" , annot=True )
plt.show()
```

```python
import statsmodels.formula.api as smf

formula = 'ACTION ~ ' + ' + '.join(list(train.drop('ACTION', axis=1).columns))

lm = smf.ols(formula=formula, data=train).fit()
lm.summary()
```

```python
new_df=pd.DataFrame()
new_df[["x1","x2","x3","x4","x5","x6","x7"]]=feature2
new_df["target"]=target2
```

```python
formula = 'target ~ ' + ' + '.join(list(new_df.drop('target', axis=1).columns))

lm = smf.ols(formula=formula, data=new_df).fit()
lm.summary()
```
