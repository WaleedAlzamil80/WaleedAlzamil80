```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.decomposition import PCA

poly=PolynomialFeatures(degree=2, include_bias=False)
pca=PCA()
```

```python
from sklearn.model_selection import train_test_split
```

```python
features=data.drop("Rings",axis=1)
target=data["Rings"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.75,random_state=99)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train.head()
```

```python
sc.fit(x_train)
```

```python
x_train=sc.transform(x_train)
x_test=sc.transform(x_test)
```

```python
poly.fit(x_train)
```

```python
x_train_poly=poly.transform(x_train)
x_test_poly=poly.transform(x_test)
```

```python
print(x_train.shape)
print(x_train_poly.shape)
```

```python
pca.fit(x_train_poly)
```

```python
var=pca.explained_variance_ratio_.cumsum()
```

```python
var
```

```python
pca8=PCA(n_components=8)
```

```python
pca8.fit(x_train_poly)
```

```python
pca8.explained_variance_ratio_.cumsum()
```

```python
xf_train=pca8.transform(x_train_poly)
xf_test=pca8.transform(x_test_poly)
```

```python
from sklearn.ensemble import AdaBoostRegressor
model=AdaBoostRegressor()
```

```python
model.fit(xf_train,y_train)
```

```python
y_pred=model.predict(xf_test)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
print(sqrt(mean_squared_error(y_test,y_pred)))
print(mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.linear_model import Ridge,ElasticNet
rid=Ridge()
```

```python
rid.fit(xf_train,y_train)
y_pred_rid=rid.predict(xf_test)
print(sqrt(mean_squared_error(y_test,y_pred_rid)))
print(mean_absolute_error(y_test,y_pred_rid))
```

```python
elas=ElasticNet()
elas.fit(xf_train,y_train)
y_pred_elas=elas.predict(xf_test)
print(sqrt(mean_squared_error(y_test,y_pred_elas)))
print(mean_absolute_error(y_test,y_pred_elas))
```

```python
from sklearn.linear_model import Lasso
lass=Lasso()
lass=Lasso()
lass.fit(xf_train,y_train)
y_pred_lass=lass.predict(xf_test)
print(sqrt(mean_squared_error(y_test,y_pred_lass)))
print(mean_absolute_error(y_test,y_pred_lass))
```

```python
from xgboost import XGBRegressor
XGBR=XGBRegressor()
XGBR.fit(xf_train,y_train)
y_pred_XGBR=XGBR.predict(xf_test)
print(sqrt(mean_squared_error(y_test,y_pred_XGBR)))
print(mean_absolute_error(y_test,y_pred_XGBR))
```

```python
from sklearn.ensemble import AdaBoostRegressor
AdaB=AdaBoostRegressor()
AdaB.fit(xf_train,y_train)
y_pred_AdaB=AdaB.predict(xf_test)
print(sqrt(mean_squared_error(y_test,y_pred_AdaB)))
print(mean_absolute_error(y_test,y_pred_AdaB))
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR=RandomForestRegressor()
RFR.fit(xf_train,y_train)
y_pred_RFR=RFR.predict(xf_test)
print(sqrt(mean_squared_error(y_test,y_pred_RFR)))
print(mean_absolute_error(y_test,y_pred_RFR))
```

#### we tried a lot of linear models and the following { RandomForestRegressor , XGBRegressor , ElasticNet  } gives us the minimal RMSE
